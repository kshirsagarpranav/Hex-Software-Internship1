```python
def fibonacci_basic(n):
    a, b = 0, 1
    series = []

    for _ in range(n):
        series.append(a)
        a, b = b, a + b

    return series

# Example
print(fibonacci_basic(10))
```

```python
def fibonacci_recursive(n):
    if n <= 1:
        return n
    return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)

# Example
for i in range(10):
    print(fibonacci_recursive(i), end=" ")
```
